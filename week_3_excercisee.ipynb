```python
'''
1. Generate a 3x4 NumPy array with random integers between 1 
and 50.   
a. Calculate and print the Mean, Median, and Standard Deviation 
of the array 
b. Print the Sum of all elements and the sum of each row. 
c. Reshape the 3x4 array into a 2x6 array and print it.
'''
import numpy as np

# part-a
arr = np.random.randint(1,50,(3,4),dtype=int)
print("The original array is:",arr)
print("\n===========\n")
mean = np.mean(arr)
median = np.median(arr)
std_dev = np.std(arr)
print("Mean: ",mean)
print("Median: ",median)
print("Standard deviation: ",std_dev)
print("\n===========\n")

#Part-b
print("The sum of all elements is: ",np.sum(arr))
print("The sum row-wise is:",arr.sum(axis=1))
print("\n===========\n")

#Part-c
reshaped = arr.reshape(2,6)
print("The reshaped array is:",reshaped)
```

```python
'''
2. Create two (3 * 3) matrices using NumPy and print it. 
Perform and print the results of the following linear algebra operations 
a. Matrix addition 
b. Matrix subtraction 
c. Matrix multiplication (element-wise and dot product) 
d. Transpose of a matrix 
e. Determinant and inverse (if applicable) 
'''
arr_1 = np.random.randint(0,10,(3,3),dtype=int)
arr_2 = np.random.randint(0,10,(3,3),dtype=int)
print("The first array is: ",arr_1)
print("")
print("The second array is: ",arr_2)
print("\n===========\n")

#Part-a
print("The sum of two matrices: ",arr_1+arr_2)
print("\n===========\n")

#Part-b
print("The difference of two matrices: ",arr_1 - arr_2)
print("\n===========\n")

#Part-c
print("Matrix multiplication:")
print("Element-wise",arr_1*arr_2)
print("")
print("Dot product:",arr_1@arr_2)
print("\n===========\n")

#Part-d
print("The transpose of the matrix arr_1 is:",arr_1.T)
print("The transpose of the matrix arr_2 is:",arr_2.T)
print("\n===========\n")

#Part-e
print("The determinant of arr_1 is:",np.linalg.det(arr_1))
print("The determinant of arr_2 is:",np.linalg.det(arr_2))
print("The inverse of arr_1 is:",np.linalg.inv(arr_1))
print("The inverse of arr_2 is:",np.linalg.inv(arr_2))
```
