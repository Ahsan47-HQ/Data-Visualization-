```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
se = pd.read_csv("sentiment.csv",index_col=0)
se
```

```python
se.drop_duplicates(inplace=True)
se.dropna(subset = ['Sentiment','Text'],inplace=True)
se
```

```python
# 1. In the sentiment feature and text, trim white spaces and convert to lowercase.
# se["Sentiment"] = se["Sentiment"].replace(se['Sentiment'].str.strip().str.lower(),inplace=True) - makes it None
# se = se.replace(se['Text'] , se["Text"].str.strip().str.lower(),inplace=True) - makes the dataframe None
se["Sentiment"] = se["Sentiment"].str.strip().str.lower()
se['Text'] = se['Text'].str.strip().str.lower()
display(se)
```

```python
# 2. What are the sentiment categories? Tabulate and visualize the number of texts for each category 
print("The Sentiment categories are:\n", se['Sentiment'].value_counts())
se_ = se.groupby("Sentiment")[["Text"]].nunique() # The column 'Text' now holds the count of texts , nunique is similar to count , but takes only unique
print("The number of texts for each category:")
se_ = se_.sort_values(by='Text',ascending=False)
display(se_)
plt.figure(figsize = (22,13))
sns.barplot(data = se_.reset_index(), x='Sentiment',y='Text',palette = 'viridis',legend=False, hue = 'Sentiment') # countplot can also be used
# .reset_index is to convert the groupby object (with an aggregated function) to a dataframe
plt.title("Count of Text grouped by sentiments")
plt.xlabel("Sentiments")
plt.ylabel("Count of Texts")
plt.xticks(rotation=90)
plt.show()
```

```python
# 3. Tabulate and Visualize the text length for each sentiment category.
se["Text_length"]=se["Text"].apply(len)
se_1 = se.groupby('Sentiment')[['Text_length']].mean()
se_1 = se_1.sort_values(by='Text_length',ascending=False)
print('The text length (mean) for each sentiment category is: ')
display(se_1)
se_1 = se_1.reset_index()
plt.figure(figsize=(22,13))
sns.barplot(data = se_1 , x = 'Sentiment' , hue = 'Sentiment', y= 'Text_length',legend=False, palette = 'viridis')
plt.title('Text Length for each sentiment category')
plt.xlabel('Sentiments')
plt.ylabel('Mean Text Length')
plt.xticks(rotation=90)
plt.show()
```

```python
# 4. Find and display the text(s) with the maximum number of likes.
display(se[se['Likes']==se['Likes'].max()][['Text','Likes']])
```

```python
# 5. Find and display the text(s) with the maximum reposts. what is its sentiment?
display(se[se['Retweets']==se['Retweets'].max()][['Text','Retweets','Sentiment']])
```

```python
# 6. Tabulate and Visualize the top 5 countries that have posted the most texts with a positive sentiment.
se['Country'] = se['Country'].str.strip()
top_5 = se[se['Sentiment']=='positive'].groupby('Country')['Text'].count().reset_index()
top_5 = top_5.sort_values(by='Text',ascending =False).head(5)
display(top_5)
sns.barplot(data = top_5, x='Country',y='Text',hue='Country',legend=False,palette='Greens_r')
plt.title("Top 5 countries with the highest amount of positive texts",fontsize=13)
plt.xlabel('Countries')
plt.ylabel('Number of positive texts')
plt.show()
```

```python
# 7. Find and display all the text(s) with the word â€˜Travelâ€™ in their hashtag.
hashes = se[se['Hashtags'].str.contains('Travel')][['Text','Hashtags']]
display(hashes)
```

```python
# 8. Find and display the most popular (most number of likes) posts (top 5) 
# from the year 2020 which were of a negative sentiment.
pops = se[(se['Sentiment']=='negative') & (se['Year']==2020)][['Text','Likes']]
pops = pops.sort_values(by = 'Likes',ascending=False).head(5)
pops
```

```python
# 9. Find and display a list of Users and their posts whose texts are labeled â€˜Happinessâ€™ sentiment.
happy = se[se['Sentiment']=='happiness'][['User','Text']]
happy
```

```python
# 10. Tabulate and visualize the years and countrywide tweets of the sentiments. Pick tweets
# with sentiments, â€˜Positive,â€™ â€˜Negative,â€™ and â€˜Neutralâ€™ only.
# s5 = s5.groupby(['Year','Country','Sentiment']).size().reset_index(name='Tweet_count') - can also be used
s5 = se[se['Sentiment'].isin(['negative','positive','neutral'])][['Year','Country','Sentiment','Text']]
display(s5)
# Thus, we can see that only for the year 2023, these sentiments were noted
req = s5.groupby('Country')[['Text','Sentiment']].count().reset_index()
req.rename(columns={'Text': 'Tweet_count'}, inplace=True)
display(req)
plt.figure(figsize = (19,12))
sns.barplot(data = req.sort_values(by='Tweet_count',ascending=False), x='Country',y= 'Tweet_count',
            hue='Country',palette = 'Blues_r',errorbar=None,legend=False)
plt.title("Country vs. Number of Texts")
plt.xlabel('Country')
plt.ylabel('Number of Texts')
plt.show()

se_f = s5.groupby(['Country','Sentiment'])['Text'].count().reset_index(name='Tweet_count')
heatmap_data = se_f.pivot(index='Country', columns='Sentiment', values='Tweet_count').fillna(0)

plt.figure(figsize=(12, 6))
sns.heatmap(heatmap_data, cmap='coolwarm', annot=True, fmt=".0f", linewidths=0.5)

plt.xlabel("Sentiment")
plt.ylabel("Country")
plt.title("Sentiment Distribution Across Countries")
plt.show()
```
